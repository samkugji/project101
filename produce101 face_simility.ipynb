유사도 함수별 측정해 보자 
* euclidean, cosine, correlation

```python
from scipy.spatial import distance

## 유사도를 측정
def most_similar(name, ds, distance_func):
    v1 = ds.loc[name]
    ret = []
    for name, row in ds.iterrows():
        ret.append( [name, round(distance_func(v1,row),5)] )
    return pd.DataFrame(ret,columns = ['name','distance']).sort_values(by=['distance'], ascending=True)
```

```python
most_similar('김세정', raw_data, distance.euclidean)
```

```python
#cosine 으로
def make_sim_matrix(distance_func):
    users = raw_data.index

    sim_matrix = []
    for u1 in users:
        similarites = []
        u1_vec = raw_data.loc[u1].values.tolist()
#              u1_vec = np.asarray(u1_vec)
#              print(u1_vec.ndim) 
        for u2 in users:
            u2_vec = raw_data.loc[u2].values.tolist()
#         u2_vec = np.asarray(u2_vec)
#         print(u2)
#         print(u2_vec.ndim,"se")
            similarites.append(round(distance_func(u1_vec, u2_vec),5))
        sim_matrix.append(similarites)

    sim_matrix = pd.DataFrame(sim_matrix, index=users, columns=users)
    return sim_matrix
```

```python
for distance_func in [distance.euclidean, distance.cosine, distance.correlation]:
    sim_matrix=make_sim_matrix(distance_func)
    print("\n=====", distance_func.__name__ , "Distance Function =====\n")
    for idx,row in sim_matrix.iterrows():
        row.sort_values(inplace=True)
        print(idx, row[1:3].index.values)
```
